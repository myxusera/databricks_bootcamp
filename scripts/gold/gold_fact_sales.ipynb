# Init

```python
import logging

logging.basicConfig(level=logging.INFO)
logger = logging.getLogger("gold.fact_sales")
```

# Transformation Logic

```python
query = """
SELECT
    sales.order_number,
    prod.product_surrogate_key,
    cust.customer_key,
    sales.order_date,
    sales.ship_date,
    sales.due_date,
    sales.sales_amount,
    sales.quantity,
    sales.price
FROM silver.crm_sales sales
LEFT JOIN gold.dim_products prod
    ON sales.product_key = prod.product_key
LEFT JOIN gold.dim_customers cust
    ON sales.customer_id = cust.customer_id;
"""
df = spark.sql(query)
```

# Write into Gold Table

```python
try:
    df.write.mode("overwrite").format("delta").saveAsTable("workspace.gold.fact_sales")
except Exception as e:
    logger.error(f"Failed to write Gold table: {e}")
    raise
```
